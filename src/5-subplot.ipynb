### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 10, 100)
y = np.sin(x)

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

x_scatter = np.random.rand(20) * 10
y_scatter = np.random.rand(20) * 10

sizes = [30, 40, 20, 10]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line Chart
axs[0, 0].plot(x, y, color='b', label='Sine Wave')
axs[0, 0].set_title('Line Chart')
axs[0, 0].legend()

# Bar Chart
axs[0, 1].bar(categories, values, color='g')
axs[0, 1].set_title('Bar Chart')

# Scatter Plot
axs[1, 0].scatter(x_scatter, y_scatter, color='r')
axs[1, 0].set_title('Scatter Plot')

# Pie Chart
axs[1, 1].pie(sizes, labels=categories, autopct='%1.1f%%', colors=['r', 'g', 'b', 'c'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = np.random.randint(1, 10, size=4)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
pie_labels = ['Apple', 'Banana', 'Cherry', 'Date']
pie_sizes = [25, 30, 15, 30]

# Create the figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart (first subplot)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

# Bar chart (second subplot)
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot (third subplot)
axs[1, 0].scatter(x_scatter, y_scatter)
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')

# Pie chart (fourth subplot)
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
