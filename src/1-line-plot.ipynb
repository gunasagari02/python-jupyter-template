### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10

# Calculate y values (square of x)
y = x**2

# Create the line plot
plt.plot(x, y, label='y = x^2')

# Add title and labels
plt.title('Line Plot of y = x^2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add grid
plt.grid(True)

# Add Legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values from 0 to 10
x = np.linspace(0, 10, 100)

# Calculate y values as the square of x
y = x**2

# Create the plot
plt.plot(x, y)

# Add title and labels
plt.title("Plot of y = x^2")
plt.xlabel("x values")
plt.ylabel("y values")

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```
